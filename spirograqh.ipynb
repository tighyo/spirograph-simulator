```python
import math
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
```

```python
d = 8
d2 = 20
rr = 2
rl = 2
l = 16

if 2*l < rr + rl + 2*d:
    print("Incorrect number!")

pi = math.pi
wr = pi / 1
wl = pi / 32
init_rr = 0
init_rl = 0

w = pi / 256
t = np.arange(0, 512, 0.01)
```

```python
xr = d + rr*np.cos(wr*t + pi/2 + init_rr)
yr = rr*np.cos(wr*t + init_rr)
xl = -d + rl*np.cos(wl*t + pi/2 + init_rl)
yl = rl*np.cos(wl*t + init_rl)
```

```python
a = 2 * (xr - xl)
b = 2 * (yr - yl)
c = (xl + xr)*(xl - xr) + (yl + yr)*(yl - yr) + (l + l)*(l - l)
```

```python
plt.scatter(xr,yr,
            c=t,
            cmap=cm.jet,
            marker='.',lw=0)
plt.scatter(xl,yl,
            c=t,
            cmap=cm.jet,
            marker='.',lw=0)
ax=plt.colorbar()
plt.show()
```

```python
D = np.abs(a*xl + b*yl + c)
numerator = np.square(a) + np.square(b)
in_sqrt = (numerator * (l**2)) - np.square(D)
```

```python
X = ((a*D - b*np.sqrt(in_sqrt)) / numerator) + xl
Y = ((b*D + a*np.sqrt(in_sqrt)) / numerator) + yl
```

```python
plt.plot(t, X, label="X")
plt.plot(t, Y, label="Y")
plt.legend()
plt.show()
```

```python
plt.scatter(np.round(X, 3), np.round(Y, 3),
            c=t,
            cmap=cm.jet,
            marker='.',lw=0)
ax=plt.colorbar()
plt.show()
```

```python
rot = w * t
rot = rot[::-1]
X_rotated = X * np.cos(rot) - (Y - d2) * np.sin(rot)
Y_rotated = X * np.sin(rot) + (Y - d2) * np.cos(rot) + d2
```

```python
plt.scatter(np.round(X_rotated, 3), np.round(Y_rotated, 3),
            c=t,
            cmap=cm.jet,
            marker='.',lw=0)
ax=plt.colorbar()
plt.show()
```
